```python
with open("initial_data.txt", encoding='utf-8') as f:
    data = f.read()
```

```python
data
```

```python
chunks = data.split("\n\n") 
chunks = [c for c in chunks if len(c)>3]
chunks
```

```python
chunks[0]
```

```python
chunks[4]
```

```python
def parse_chunk(chunk): 
        chunk = chunk.strip()
        lines = chunk.split("\n")
        name = lines[0].replace("Name: ", "")
        posts = lines[1].replace("Posts: ", "")
        followers = lines[2].replace("Followers: ", "")
        following = lines[3].replace("Following: ", "")
        bio = lines[4].replace("Bio: ", "")
        type_of_page = lines[4].replace("type_of_page: ", "")
        
        return {"name": name, "posts": posts, "followers": followers, "following": following,"bio": bio,"type_of_page": type_of_page}

#print(),(parse_chunk(chunks[0]))
```

```python
all_chunks = [] 
for chunk in chunks:  
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(all_chunks)
```

```python
import json
s = json.dumps(all_chunks, indent=4)
with open("data.json", "w") as f:
    f.write(s)
```

```python
all_data = []
for c in chunks:
    all_data.append(parse_chunk(c))
all_data
```

# Who has maximum posts ?

```python
max_posts = 0
for chunk in all_chunks:
    posts = int(chunk['posts'].split()[0].replace(',','')) #Remove commas
  #extract number and convert to int
    
    if max_posts < posts:
        max_posts = posts
        chunk_with_max_post = chunk
print(chunk_with_max_post)
        
```

# Who has maximum Follwers ?

```python
max_followers = 0
for chunk in all_chunks:
    if max_followers < followers:
        max_followers = followers
        chunk_with_max_followers = chunk
print(chunk_with_max_followers)
        
```

# Who follows maximum people ?

```python
max_following = 0
chunk_with_max_following = None

for chunk in all_chunks:
    following = int(chunk['following'].split()[0].replace(',',''))
    
    if max_following < following:
        max_following = following
        chunk_with_max_following = chunk
print(chunk_with_max_following)
        
```
